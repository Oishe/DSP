\begin{equation*}
e^{j\theta } =cos( \theta ) \ +\ jsin( \theta )
\end{equation*}


\begin{gather*}
x_{p}( t) =a_{0} \ +\ 2\sum ^{\infty }_{n\ =\ 1}\Bigl[ a_{n} \ cos\left(\tfrac{2\pi nt}{T_{0}}\right) +b_{n} \ sin\left(\tfrac{2\pi nt}{T_{0}}\right)\Bigr]\\
f_{0} \ =\ \tfrac{1}{T_{0}} \ ;\ fundamental\ frequency\\
\tfrac{n}{T_{0}} \ \ ;\ n^{th} \ harmonic
\end{gather*}


\begin{gather*}
cos( \theta ) \ =\ \frac{1}{2}\Bigl[ e^{j\theta } +e^{-j\theta }\Bigr] ;\ sin( \theta ) \ =\ \frac{1}{2j}\Bigl[ e^{j\theta } -e^{-j\theta }\Bigr]\\
a_{n} cos( 2\pi nt/T_{0}) \ =\ \frac{a_{n}}{2}\Bigl[ e^{j2\pi nt/T_{0}} +e^{-j2\pi nt/T_{0}}\Bigr]\\
b_{n} sin( 2\pi nt/T_{0}) \ =\ -j\frac{b_{n}}{2}\Bigl[ e^{j2\pi nt/T_{0}} -e^{-j2\pi nt/T_{0}}\Bigr]\\
a_{n} cos( 2\pi nt/T_{0}) \ +\ \ b_{n} sin( 2\pi nt/T_{0}) \ =\ \frac{1}{2}\Bigl[( a_{n} -jb_{n}) e^{j2\pi nt/T_{0}} +( a_{n} +jb_{n}) e^{-j2\pi nt/T_{0}}\Bigr]\\
( a_{n} -jb_{n})^{*} \ =\ ( a_{n} +jb_{n}) \ \text{complex conjugates}\\
c_{n} \ =\ \begin{cases}
a_{n} -jb_{n} & n >0\\
a_{0} & n=0\\
a_{n} +jb_{n} & n< 0
\end{cases}\\
x_{p}( t) \ =\ \sum ^{\infty }_{n=-\infty } c_{n} e^{j2\pi nt/T_{0}}\\
c_{n} \ =\ \frac{1}{T_{0}}\int ^{T_{0} /2}_{-T_{0} /2} x_{p}( t) e^{-j2\pi nt/T_{0}} dt
\end{gather*}


\begin{gather*}
x_{p}( t) \ =\ \begin{cases}
1 & -1< t< 1\\
0 & \text{otherwise}
\end{cases} \ T_{0} =4\\
c_{n} =\frac{1}{4}\int ^{-1}_{1}( 1) e^{-j2\pi nt/4} dt\\
c_{n} =\frac{-1}{2j\pi n}\Bigl[ e^{-j\pi nt/2}\Bigr]^{t=1}_{t=-1}\\
c_{n} =\frac{1}{2j( \pi n)}\Bigl[ e^{j\pi n/2} -e^{-j\pi n/2}\Bigr]\\
c_{n} =\frac{1}{( \pi n)} sin( \pi n/2)\\
c_{n} =\frac{1}{2}\frac{sin( \pi n/2)}{\pi n/2}\\
\end{gather*}


\begin{gather*}

\end{gather*}

```python
import numpy as np
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True) 
```

```python
n = 100
x = np.arange(-2, 2, 0.01)

# # without using complex
# y = 0.5*np.ones(x.shape)
# for k in range(1, n+1):
#     y = y + (np.sin(k*np.pi/2)/(k*np.pi/2))*np.cos(x*k*np.pi/2)

# using complex
y = np.zeros(x.shape, dtype=np.complex_)
for k in range(-n, n+1):
    if k == 0:
        y = y + 0.5*np.ones(x.shape, dtype=np.complex_)
    else:
        y = y + 0.5*(np.sin(k*np.pi/2)/(k*np.pi/2))*np.exp(1j*x*k*np.pi/2)
y = y.real

trace0 = go.Scatter(
    x=[-2, -1, -1, 0, 1, 1, 2],
    y=[0, 0, 1, 1, 1, 0, 0]
)
trace1 = go.Scatter(
    x=x.tolist(),
    y=y.tolist()
)
data = [trace0, trace1]

iplot(data)
```
